```python
import pandas as pd
import re

# Load the Excel file
excel_file_path = ('C:/Users/HOME/PycharmProjects/Internship/cleaned_file.xlsx')  # Replace with the actual file path
df = pd.read_excel(excel_file_path)
# Convert all string-type columns to lowercase
df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)

# Specify the column index (index 4 corresponds to the fifth column, as indexing starts from 0)
column_index = 4

# Extract numeric values using regular expression
df.iloc[:, column_index] = df.iloc[:, column_index].astype(str)  # Convert column to string
df.iloc[:, column_index] = df.iloc[:, column_index].apply(lambda x: [int(match.group()) for match in re.finditer(r'\d+', x)] if pd.notnull(x) else [])

# Calculate the mean and update the original column
df.iloc[:, column_index] = df.iloc[:, column_index].apply(lambda x: int(sum(x) / len(x)) if len(x) > 0 else None)

# Display the updated DataFrame
df

# Get the directory of the original file
original_directory = os.path.dirname('C:/Users/HOME/PycharmProjects/Internship/cleaned_file.xlsx')

# Save the updated DataFrame to the original Excel file
df.to_excel(excel_file_path, index=False)

# Display a message indicating the update
print(f"Updated file saved at: {excel_file_path}")
```

```python
import pandas as pd
import numpy as np

# Load the Excel file into a DataFrame
df = pd.read_excel('C:/Users/HOME/PycharmProjects/Internship/cleaned_file.xlsx')

# Assume the column with outliers is named 'your_outlier_column'
your_outlier_column = df.loc[4]

# Calculate the median and interquartile range (IQR)
median_value = your_outlier_column.median()
Q1 = your_outlier_column.quantile(0.25)
Q3 = your_outlier_column.quantile(0.75)
IQR = Q3 - Q1

# Define a threshold for outliers (e.g., 1.5 times the IQR)
outlier_threshold = 1.5 * IQR

# Identify outliers using the threshold
outliers = (your_outlier_column < Q1 - outlier_threshold) | (your_outlier_column > Q3 + outlier_threshold)

# Replace outliers with the median value
your_outlier_column[outliers] = median_value

# Update the DataFrame with the modified column
df['your_outlier_column'] = your_outlier_column
df

# Now, your DataFrame has outliers replaced with the median value in the specified column
```
